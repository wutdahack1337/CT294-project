```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("energydata_complete.csv")
df
```

```python
training, testing = train_test_split(df, test_size=0.25, random_state=1337)
```

```python
training = training.sort_values(by="date").reset_index(drop=True)
training = training.iloc[:,:-2]
training.to_csv("training.csv", index=False)
training
```

```python
testing = testing.sort_values(by="date").reset_index(drop=True)
testing = testing.iloc[:, :-2]
testing.to_csv("testing.csv", index=True)
testing
```
