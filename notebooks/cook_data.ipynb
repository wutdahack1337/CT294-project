```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("datasets/energydata_complete.csv")
df["date"] = pd.to_datetime(df["date"])
df["Appliances"] = pd.to_numeric(df["Appliances"])
df.set_index("date", inplace=True)
df
```

```python
training, testing = train_test_split(df, test_size=0.25, random_state=1337)

training = training.sort_values(by="date")
training = training.iloc[:,:-2]
training.to_csv("datasets/training.csv")

testing = testing.sort_values(by="date")
testing = testing.iloc[:, :-2]
testing.to_csv("datasets/testing.csv")
```
