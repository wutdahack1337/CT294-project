```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("energydata_complete.csv")
df["date"] = pd.to_datetime(df["date"])
df["Appliances"] = pd.to_numeric(df["Appliances"])
df.set_index("date", inplace=True)
df
```

```python
plt.figure(figsize=(16, 9))
plt.xlabel("Thá»i gian", fontsize=16)
plt.xticks(fontsize=13)
plt.ylabel("NÄƒng lÆ°á»£ng tiÃªu thá»¥ (Wh)", fontsize=16)
plt.yticks(fontsize=13)
plt.grid(True)

plt.plot(df.index, df["Appliances"], linewidth=0.3)
plt.savefig("fig1.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
firstweek_df = df["2016-01-11":"2016-01-18"]

plt.figure(figsize=(16, 9))
plt.xlabel("Thá»i gian", fontsize=16)
plt.xticks(fontsize=13)
plt.ylabel("NÄƒng lÆ°á»£ng tiÃªu thá»¥ (Wh)", fontsize=16)
plt.yticks(fontsize=13)
plt.grid(True)

plt.plot(firstweek_df.index, firstweek_df["Appliances"], linewidth=0.9)
plt.savefig("fig2.png", dpi=300, bbox_inches="tight")

plt.show()
```
